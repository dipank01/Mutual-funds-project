```python
import pandas as pd
import numpy
import yfinance as yf
```

```python
import yfinance as yf

# Mapping of categories to tickers
category_tickers = {
    "LargeCap": ["^NSEI", "^BSESN", "RELIANCE.NS", "HDFCBANK.NS", "ICICIBANK.NS", "INFY.NS", "TCS.NS"],
    "MidCap": ["^NSEMDCP50", "MAXHEALTH.NS", "TIINDIA.NS", "BHARATFORG.NS", "AUBANK.NS", "ZYDUSLIFE.NS"],
    "SmallCap": ["^CNXSC", "CDSL.NS", "KPITTECH.NS", "TATAELXSI.NS", "EQUITASBNK.NS", "CYIENT.NS"]
}

# Flatten list of tickers
all_tickers = list(set(sum(category_tickers.values(), [])))



# Download from yfinance
data = yf.download(all_tickers, start="2015-01-01", end="2025-08-01", group_by='ticker')

print(data.head())
```

```python
df_long = data.stack(level=[0], future_stack=True)
df_long
```

```python
# Flatten the index into columns
df_long = df_long.reset_index()

# Rename for consistency
df_long = df_long.rename(columns={
    'Date': 'date',
    'Ticker': 'ticker',
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'Adj Close': 'adj_close',
    'Volume': 'volume'
})

```

```python
df_long.drop(columns=['adj_close'], inplace=True)
```

```python
df_long.dropna(inplace=True)
df_long
```

```python
# Save to CSV
df_long.to_csv( 'all_tickers_data.csv', index=False)
```

```python
tickers_data = {}

for ticker in all_tickers:
    ticker_object = yf.Ticker(ticker)

    #convert info() output from dictionary to dataframe
    temp = pd.DataFrame.from_dict(ticker_object.info, orient="index")
    temp.reset_index(inplace=True)
    temp.columns = ["Attribute", "Recent"]
    
    # add (ticker, dataframe) to main dictionary
    tickers_data[ticker] = temp
    
# Combine all dataframes into a single dataframe

combined_data = pd.concat(tickers_data)
combined_data = combined_data.reset_index()
combined_data


```

```python
del combined_data["level_1"] # clean up unnecessary column
combined_data.columns = ["Ticker", "Attribute", "Recent"] # update column names

combined_data
```

```python
hash_count = {}

for attr in combined_data['Attribute']:
    if attr in hash_count:
        hash_count[attr] += 1
    else:
        hash_count[attr] = 1

print(combined_data['Ticker'].nunique())
print(hash_count)
```

```python
hash_count.keys()
```

```python
common_attr = []
for attr in hash_count.keys():
    if hash_count[attr] == 19:
        common_attr.append(attr)
            
print("Common attributes across all tickers:")
print(common_attr)
```

```python
attr_selected = ['currency', 'symbol', 'region', 'typeDisp', 'shortName', 'longName', 'exchange', 'fullExchangeName', 'exchangeTimezoneName','exchangeTimezoneShortName', 'gmtOffSetMilliseconds', 'market', 'quoteType',
                  'marketState', 'gmtOffSetMilliseconds']
    

filtered_data = combined_data[combined_data['Attribute'].isin(attr_selected)]
print(filtered_data)
```
